```python
# O dataframe de cada banco é especificado pelo seu ticker. Vamos começar com o Bank of America
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
BAC = data.DataReader ("BAC", "yahoo", start, end)
```

```python
BAC
```

```python
# Vamos ver a série de fechamentos do Bank of America para notar como este banco reagiu às crises
# da década de 2006/01/01-2016/01/01
BAC['Close'].plot()
```

#### Agora vamos ler os dataframes dos outros bancos.

```python
# CityGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

#### Para analisarmos todos esses DataFrames juntos de uma vez vamos primeiro criar uma lista com todos eles em ordem alfabética

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

#### Agora vamos usar pd.concat para concatenar os DataFrames dos bancos juntos em um único chamado bank_stocks. Defina o argumento das chaves igual à lista de tickers. Também preste atenção em que eixo você concatena.

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys= tickers)
bank_stocks.head(6)
```

```python
# Vamos agora nomear os índices que especificam os dados desse DataFrame conjunto
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

#### Crie um novo DataFrame vazio chamado returns. Este dataframe conterá os retornos para o ação de cada banco. Os retornos geralmente são definidos por:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

### Primeiro vejamos o método percentage change "pct_change", o qual é bem usado para calcular diferenças relativos de elementos numa dada série.

```python
df1 = pd.DataFrame({

    'FR': [4.0405, 4.0963, 4.3149],

    'GR': [1.7246, 1.7482, 1.8519],

    'IT': [804.74, 810.01, 860.13]},

    index=['1980-01-01', '1980-02-01', '1980-03-01'])
df1.pct_change()
```

```python
returns = pd.DataFrame()
```

#### Podemos agora usar o método pct_change() do pandas na coluna Close para criar uma coluna que represente esse valor de retorno. Criaremos um loop for que vá para cada Bank Stock Ticker, cria essa coluna de retorno e configura-a como uma coluna nos dados DataFrame. 

```python
for ticker in tickers:
    returns[ticker + ' Return'] = bank_stocks[ticker]['Close'].pct_change()
returns.head()    
```

#### Dê uma olhada no desvio padrão dos retornos. Qual ação você classificaria como a mais arriscada durante todo o período de tempo? Qual você classificaria como a mais arriscado para o ano 2015? 


### Observação: o desvio padrão é uma métrica que é normalmente utilizada no mercado financeiro para medir o grau de risco de um ativo. Ao medir o desvio padrão de retorno por retorno podemos ver o quão grande é a variação desses retornos e, com isso, podemos ver o quão perigoso e imprevisível é o ativo, pois quanto mais volátil ele for mais perigoso e imprevisível será o ativo.

```python
# Vejamos os desvios padrões de toda a śerie returns
returns.std()
```

## Médias móveis

Vamos analisar as médias móveis para essas ações no ano de 2008.

### Trace a média de 30 dias para o preço próximo do Bank Of America para o ano de 2008 

```python
BAC['Close'].head(5)
```

```python
BAC2008 = BAC['Close'][BAC.index.year == 2008]
BAC2008
```

```python
# Método Rolling                     Média com 10
BAC2008_media_movel_10 = BAC2008.rolling(window = 10).mean()
BAC2008_media_movel_10
```

```python
BAC2008.rolling(window = 10).mean().plot()
```

```python
# Média com 30
BAC2008_media_movel_30 = BAC2008.rolling(window = 30).mean().plot()
```

```python
plt.figure(figsize=(12, 6))
BAC2008.rolling(window = 30).mean().plot()
```

```python
BAC2008.plot()
```

```python
plt.figure(figsize=(12, 6))
BAC2008.rolling(window = 30).mean().plot()
BAC2008.plot()
```

```python
plt.figure(figsize=(12, 6))
BAC2008.rolling(window = 30).mean().plot(label='Média móvel de 30 dias')
BAC2008.plot(label='BAC Fechamento de 2008')
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
BAC2008.rolling(window = 30).mean().plot(label='30 day average')
BAC2008.plot(label='BAC Close')
plt.legend()
```
