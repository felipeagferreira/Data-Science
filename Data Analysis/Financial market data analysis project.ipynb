```python
# Each bank's data frame is specified by its ticker. Let's start with Bank of America
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
BAC = data.DataReader ("BAC", "yahoo", start, end)
```

```python
BAC.head()
```

```python
BAC.info()
```

```python
# Let's look at Bank of America's series of closings to see how this bank reacted to the crises
# of the decade of 01/01/2006-01/01/2016.
BAC['Close'].plot()
```

```python
# Now let's read the dataframes of the other banks.


# CityGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

#### Now we are going to use pd.concat to concatenate the banks' Data Frames together into a single one called "bank_stocks". Set the key argument to the ticker list.

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys= tickers)
bank_stocks.head(6)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

#### Let's create a new empty Data Frame called returns. This dataframe will contain the returns for each bank's stock. Returns are generally defined by:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

#### First let's look at the percentage change method "pct_change", which is well used to calculate relative differences of elements in a given series.

```python
df1 = pd.DataFrame({

    'FR': [4.0405, 4.0963, 4.3149],

    'GR': [1.7246, 1.7482, 1.8519],

    'IT': [804.74, 810.01, 860.13]},

    index=['1980-01-01', '1980-02-01', '1980-03-01'])
df1.pct_change()
```

```python
returns = pd.DataFrame()
```

#### We can now use the pandas pct_change () method in the Close column to create a column that represents this return value. We will create a for loop that goes to each Bank Stock Ticker, creates that return column and sets it up as a column in the DataFrame data.

```python
for ticker in tickers:
    returns[ticker + ' Return'] = bank_stocks[ticker]['Close'].pct_change()
returns.head()    
```
